```python
fig,ax = plt.subplots()
# El valor más grande
valor_maximo = np.max(datos)
print("El valor más grande:", valor_maximo)

# El valor más pequeño
valor_minimo = np.min(datos)
print("El valor más pequeño:", valor_minimo)

# Obtuvieron 125 o más
obtuvieron_125_o_mas = np.mean(datos >= 125) * 100
print("Porcentaje de estudiantes que obtuvieron 125 o más:", obtuvieron_125_o_mas)

# Obtuvieron 138 o menos
obtuvieron_138_o_menos = np.mean(datos <= 138) * 100
print("Porcentaje de estudiantes que obtuvieron 138 o menos:", obtuvieron_138_o_menos)

# Porcentaje de valores entre 129 y 146
porcentaje_entre_129_146 = np.mean((datos >= 129) & (datos <= 146)) * 100
print("Porcentaje de valores entre 129 y 146:", porcentaje_entre_129_146)

# Frecuencias absolutas simples
frecuencias_absolutas = np.bincount(datos)
print("Frecuencias absolutas:", frecuencias_absolutas)

# Frecuencias absolutas acumuladas
frecuencias_absolutas_acumuladas = np.cumsum(frecuencias_absolutas)
print("Frecuencias absolutas acumuladas:", frecuencias_absolutas_acumuladas)

# Frecuencias relativas simples
frecuencias_relativas = frecuencias_absolutas / len(datos)
print("Frecuencias relativas:", frecuencias_relativas)

# Frecuencias relativas acumuladas
frecuencias_relativas_acumuladas = np.cumsum(frecuencias_relativas)
print("Frecuencias relativas acumuladas:", frecuencias_relativas_acumuladas)

# Media aritmética de valores agrupados
media_aritmetica_agrupados = np.mean(datos)
print("Media aritmética de valores agrupados:", media_aritmetica_agrupados)

# Media aritmética de valores no agrupados
media_aritmetica_no_agrupados = statistics.mean(datos)
print("Media aritmética de valores no agrupados:", media_aritmetica_no_agrupados)

# Media geométrica
media_geometrica = statistics.geometric_mean(datos)
print("Media geométrica:", media_geometrica)

# Media armónica
media_armonica = statistics.harmonic_mean(datos)
print("Media armónica:", media_armonica)

# Mediana
mediana = np.median(datos)
print("Mediana:", mediana)

# Moda
moda = statistics.mode(datos)
print("Moda:", moda)

# Múltiples modas
modas = statistics.multimode(datos)
print("Múltiples modas:", modas)

# Cuartiles
cuartiles = np.percentile(datos, [25, 50, 75])
print("Cuartiles:", cuartiles)

# Deciles
deciles = np.percentile(datos, [10, 20, 30, 40, 50, 60, 70, 80, 90])
print("Deciles:", deciles)

# Percentiles
percentiles = np.percentile(datos, [25, 50, 75, 90])
print("Percentiles:", percentiles)

# Histograma
plt.hist(datos, bins=10, edgecolor='black')
plt.xlabel("Pesos")
plt.ylabel("Frecuencia")
plt.title("Histograma de Pesos de Estudiantes")
ax.grid()
plt.show()
```
