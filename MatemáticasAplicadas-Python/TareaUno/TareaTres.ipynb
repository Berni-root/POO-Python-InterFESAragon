```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
x = np.linspace(-10, 10, 100)
y1 = (6 - 3*x) / 2
y2 = (-9 - 6*x) / 4

plt.plot(x, y1, label='3x + 2y = 6')
plt.plot(x, y2, label='6x + 4y = -9')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Gr√°fico de las ecuaciones')
plt.legend()
plt.grid(True)
plt.show()
```

```python
x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)
z = np.linspace(-10, 10, 100)

X, Y, Z = np.meshgrid(x, y, z)

eq1 = 2*X - Y + 2*Z - 3
eq2 = 2*X + 3*Y - Z
eq3 = X + 2*Y + 2*Z + 12

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.plot_surface(X, Y, Z, facecolors='r', alpha=0.5, linewidth=0, antialiased=False, rstride=1, cstride=1)
ax.plot_surface(X, Y, eq2, facecolors='g', alpha=0.5, linewidth=0, antialiased=False, rstride=1, cstride=1)
ax.plot_surface(X, eq3, Z, facecolors='b', alpha=0.5, linewidth=0, antialiased=False, rstride=1, cstride=1)

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.legend(['2x - y + 2z = 3', '2x + 3y - z = 0', 'x + 2y + 2z = -12'])

plt.show()
```

```python
# Definir la funci√≥n de la par√°bola
def parabola(x):
    return x**2 + 3

# Definir el punto de tangencia
x0 = 1
y0 = 4

# Definir el rango de valores de x para graficar
x = np.linspace(-10, 10, 100)

# Calcular la pendiente de la recta tangente utilizando la derivada de la funci√≥n
# La derivada de y = x^2 + 3 es dy/dx = 2x
pendiente = 2 * x0

# Calcular el desplazamiento vertical (ordenada al origen) de la recta tangente
ordenada_origen = y0 - pendiente * x0

# Calcular los valores y de la recta tangente
y_tangente = pendiente * x + ordenada_origen

# Graficar la par√°bola y la recta tangente
plt.plot(x, parabola(x), label='Par√°bola')
plt.plot(x, y_tangente, label='Recta tangente')

# Graficar el punto de tangencia
plt.plot(x0, y0, 'ro', label='Punto de tangencia')

# Etiquetas de los ejes y t√≠tulo del gr√°fico
plt.xlabel('x')
plt.ylabel('y')
plt.title('Recta tangente a la par√°bola')

# Mostrar la leyenda
plt.legend()

# Mostrar el gr√°fico
plt.show()
```

```python
#funci√≥n de la par√°bola
def parabola(x):
    return x**2 + 3

#objetivo
objetivo_x = 0.5
objetivo_y = 0
punto_impacto_x = objetivo_x
punto_impacto_y = parabola(objetivo_x)

x = np.linspace(-10, 10, 100)
y_parabola = parabola(x)

plt.plot(x, y_parabola, label='Par√°bola')
plt.plot(objetivo_x, objetivo_y, 'ro', label='Objetivo')
plt.plot(punto_impacto_x, punto_impacto_y, 'go', label='Punto de impacto')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Punto de impacto del proyectil')
plt.legend()
plt.show()

print(f'El punto de impacto del proyectil es ({punto_impacto_x}, {punto_impacto_y})')
```
